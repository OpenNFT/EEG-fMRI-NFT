```python
import time

import requests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
times = []

for _ in range(1000):
    start = time.time()
    internal_time = requests.get('http://127.0.0.1:8000/get_prediction_time')
    end = time.time()
    duration = end - start
    times.append(
        {'Inference time': float(internal_time.content),
         'Request time': duration}
    )
    time.sleep(0.01)
times = pd.DataFrame(times)
```

```python
times
```

```python
sns.violinplot(times)
plt.ylabel('Time, seconds')
plt.title('Inference and total request time, 1000 replicas')
```
